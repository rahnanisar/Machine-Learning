```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("50_Startups.csv")
```

```python
data.head()
```

```python
data["State"].value_counts()
```

```python
#GETTING THE DUMMIES FOR STATE COLUMN
data1 = pd.get_dummies(data["State"])
data1.head()
```

```python
#concatinating the data frames
data = pd.concat([data,data1],axis=1)
data.head()
```

```python
data.drop(["State"],axis=1,inplace=True)
```

```python
x=data.drop(["Profit"],axis=1)
x
```

```python
y=data["Profit"]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train) #training
```

```python
y_pred=model.predict(x_test)   #testing
y_pred
```

```python
df=pd.DataFrame(data={"Predicted value":y_pred,"Actual value":y_test})
df
```

```python
model.intercept_
```

```python
model.coef_
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
model.score(x_test,y_test) #testing data accuracy score
```

```python
model.score(x_train,y_train)  #training data  model accuracy score
```

```python
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(y_test,y_pred))
```
