```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("horror_movies.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isna().sum()
```

```python
data['release_date']=pd.to_datetime(data['release_date'])
```

```python
data.info()
```

```python
data['release_date'].value_counts().head(10).plot(kind='bar')    #plot.bar can also be given
```

```python
sub_data=data[['genre_names','title','budget']].sort_values(by='budget',ascending=False).head(10)
plt.figure(figsize=(15,6))
sns.barplot(data=sub_data,x='title',y='budget')
plt.xticks(rotation=90)
plt.show()
```

```python
lowest=data[['genre_names','budget','title']].sort_values(by='budget').tail(10)
plt.figure(figsize=(15,5))
sns.barplot(data=lowest,x='genre_names',y='budget')
plt.xticks(rotation=90)
plt.show()
```
