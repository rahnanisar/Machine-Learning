```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("Real estate.csv")
data.head()
```

```python
data.info()
```

```python
x=data.drop("Y house price of unit area",axis=1)
y=data["Y house price of unit area"]
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import PolynomialFeatures
model=PolynomialFeatures(degree=3)
poly_features=model.fit_transform(x)
poly_features.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(poly_features,y,test_size=30,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
polymodel=LinearRegression()
polymodel.fit(x_train,y_train)
```

```python
y_pred=polymodel.predict(x_test)
```

```python
pd.DataFrame({'Y_Test':y_test,'Y_pred':y_pred,'Residuals':(y_test-y_pred)}).head(5)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
poly_MAE=mean_absolute_error(y_test,y_pred)
poly_MSE=mean_squared_error(y_test,y_pred)
poly_RMSE=np.sqrt(poly_MSE)
pd.DataFrame([poly_MAE,poly_MSE,poly_RMSE],
             index=['m','me','RMSE'],columns=['metrics'])
```
