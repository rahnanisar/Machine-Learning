```python
from sklearn.impute import SimpleImputer
si=SimpleImputer(strategy="mean")     #meadian,most_frequent

arr=si.fit_transform(data[['CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term',
       'Credit_History']])
arr
```

```python
new=pd.DataFrame(arr,columns=data.select_dtypes(include="float64").columns)
new
```

```python
new.isna().sum()
```

```python
data.drop(['CoapplicantIncome','LoanAmount','Loan_Amount_Term','Credit_History'],axis=1,inplace=True)
data
```

```python
newdata=pd.concat([data,new],axis=1)
newdata
```
