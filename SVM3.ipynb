```python
data=pd.read_csv("drug.csv")
data.head()
```

```python
data.replace({'BP':{'LOW':0,'NORMAL':1,'HIGH':2}},inplace=True)
```

```python
data.replace({'Sex':{'F':0,'M':1}},inplace=True)
```

```python
data.replace({'Cholesterol':{'HIGH':0,'NORMAL':1}},inplace=True)
```

```python
x=data.drop(["Drug"],axis=1)
y=data["Drug"]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=45,test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
#from sklearn.svm import svc
from sklearn import svm
clf=svm.SVC(kernel='linear')
clf. fit(x_train,y_train)
y_pred=clf.predict(x_test)
```

```python
from sklearn import metrics 
print("Accuracy:",metrics.accuracy_score(y_test,y_pred))
```

```python
from sklearn import svm
#create a sum classifier
clf2=svm.SVC(kernel='rbf')
#train the model using the training sets
clf2. fit(x_train, y_train)
#predict the response for datasetr 
y_pred2=clf2.predict(x_test)
```

```python
from sklearn import metrics 
print("Accuracy:",metrics.accuracy_score(y_test,y_pred2))
```

```python
from sklearn import svm
#create a sum classifier
clf1=svm.SVC(kernel='poly')
#train the model using the training sets
clf1. fit(x_train, y_train)
#predict the response for datasetr 
y_pred1=clf1.predict(x_test)
```

```python
from sklearn import metrics 
print("Accuracy:",metrics.accuracy_score(y_test,y_pred1))
```

```python
from sklearn import svm
#create a sum classifier
clf3=svm.SVC(kernel='sigmoid')
#train the model using the training sets
clf3. fit(x_train, y_train)
#predict the response for datasetr 
y_pred3=clf3.predict(x_test)
```

```python
from sklearn import metrics 
print("Accuracy:",metrics.accuracy_score(y_test,y_pred3))
```
