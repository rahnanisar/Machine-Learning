```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("movies.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data['run_time']=data['run_time'].str.replace('min','')
data['gross_total']=data['gross_total'].str.replace('M','')
data['gross_total']=data['gross_total'].str.replace('$','')
data['votes']=data['votes'].str.replace(',','')
```

```python
data.head()
```

```python
data['run_time']=data['run_time'].astype('int64')
data['gross_total']=data['gross_total'].astype('float64')
data['votes']=data['votes'].astype('int64')
```

```python
data.info()
```

```python
data["category"].value_counts().plot(title='most common category',kind='pie',autopct='%1.2f',cmap='plasma')
plt.show()
```

```python
data["genre"].value_counts().head().plot(title='most common genre',kind='pie',autopct='%1.2f',cmap='rainbow')
plt.show()
```

```python
bottom=data[['gross_total','movie_name']].sort_values(ascending=False,by='gross_total').tail()
plt.figure(figsize=(20,10))
sns.barplot(data=bottom,x='movie_name',y='gross_total',color='black')
plt.show()
           
                                                    
                                                       
```

```python
bottom=data[['imdb_rating','movie_name']].sort_values(ascending=False,by='imdb_rating').tail()
plt.figure(figsize=(20,10))
sns.barplot(data=bottom,x='movie_name',y='imdb_rating',color='m')
plt.show()
           
                                                    
```
