#### scaling

```python
#its a method to reduce size of a perticular value in a given data without changing orginal value
```

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("loan.csv")
```

```python
data.head()
```

#### standard scaling

###### standization/scaling
#### (value-mean)/standard deviation

```python
data["ApplicantIncome"]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.title("before")
sns.distplot(data["ApplicantIncome"])
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
ss.fit(data[["ApplicantIncome"]])
data["Applicant"]=ss.transform(data[["ApplicantIncome"]])
```

```python
data["Applicant"]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.title("after")
sns.distplot(data["Applicant"])
```

######## normalization min_max Scaling

```python
####### (value_min)(max)_(min)
```

```python
from sklearn.preprocessing import MinMaxScaler
ms=MinMaxScaler()
ms.fit(data[["CoapplicantIncome"]])
data["Coapplicants"]=ms.transform(data[["CoapplicantIncome"]])
```

```python
data["Coapplicants"]
```

```python
plt.subplot(1,2,1)
plt.title("before")
sns.distplot(data["CoapplicantIncome"])
plt.show()
```

```python
plt.subplot(1,2,1)
plt.title("after")
sns.distplot(data["Coapplicants"])
plt.show()
```

#### Robust scaling
###### (value_median)/IQR (inter quartile range =IQR=q3-q1

```python
from sklearn.preprocessing import RobustScaler
ft=RobustScaler()
ft.fit(data[["CoapplicantIncome"]])
data["Coapplicant"]=ft.transform(data[["CoapplicantIncome"]])
```

```python
data["Coapplicant"]
```

```python
plt.subplot(1,2,1)
plt.title("before")
sns.distplot(data["CoapplicantIncome"])
plt.show()
```

```python
plt.subplot(1,2,1)
sns.distplot(data['Coapplicant'])
plt.title("After")
plt.show()
```
