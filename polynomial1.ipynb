```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("Position_Salaries.csv")
data.head()
```

```python
plt.scatter(data=data,x="Level",y="Salary")
plt.show()
```

```python
#x=data["Level"]
#y=data["Salary"]
x=data.iloc[:,1:-1].values
y=data.iloc[:,-1].values
```

```python
x
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
linear=LinearRegression()
linear.fit(x,y) #training
```

```python
linear.predict(x)
```

```python
plt.scatter(x,y,color="red")
plt.plot(x,linear.predict(x))
plt.title("Using Linear regression")
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
classifier=PolynomialFeatures(degree=2)
x_poly=classifier.fit_transform(x)
x_poly
```

```python
linear=LinearRegression()
linear.fit(x_poly,y)
```

```python
linear.predict(x_poly)
```

```python
plt.scatter(x,y,color="red")
plt.plot(x,linear.predict(x_poly))
plt.title("Using Linear regression")
plt.show()
```
