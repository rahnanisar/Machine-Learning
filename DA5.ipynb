```python
data['InvoiceDate']=pd.to_datetime(data['InvoiceDate'])
```

```python
data.duplicated().sum()
```

```python
data['month']=data['InvoiceDate'].dt.month
data['year']=data['InvoiceDate'].dt.year
data['date']=data['InvoiceDate'].dt.date
```

```python
data['date']=pd.to_datetime(data['date'])
```

```python
data['Total_Sales'] = data['Quantity']*data['UnitPrice']
```

```python
data.info()
```

```python
#count year basis sales calculation

plt.figure(figsize=(15,5))
sns.barplot(x='year',y='Total_Sales',data=data,color='r')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='month',y='Total_Sales',hue='year',data=data,color='r')
plt.show()
```

```python
n=data.groupby('Description')['Total_Sales'].value_counts(ascending=False).head().plot()
plt.xticks(rotation=90)
plt.show()
```
