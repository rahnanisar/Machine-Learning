```python
#To remove unwanted cell we can drop it (here we have index column)
data.drop(["index"],axis=1,inplace=True)
data
```

```python
data.duplicated().sum()
```

```python
data.isnull().sum()
```

```python
data['age'].fillna(data['age'].mean(),inplace=True)
```

```python
data['region'].fillna(data['region'].mode()[0],inplace=True)
```

```python
data.isnull().sum()
```

```python
sns.countplot(x="smoker",data=data)
plt.show()
```
